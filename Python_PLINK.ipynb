## PLINK Class

```python
class PLINK: 
    def __init__(self):
        
        print("Hello PLINK")
        
    def assoc(self, gwas_file_name, pheno , allow_no_sex = True ,output_name="result"):
        fid = pd.DataFrame(pheno.iloc[:,0])
        new_pheno = pd.concat([fid,pheno],1)
        
        if (allow_no_sex) == True:
            allow_no_sex = "--allow-no-sex"
        
        else:
            allow_no_sex = None
            
        
        if sum(new_pheno.columns == ['FID','IID','Y']) == 3:
            
            new_pheno.loc[(new_pheno['Y']==1),"Y"]=2
            new_pheno.loc[(new_pheno['Y']==0),"Y"]=1
                
            
            new_pheno.to_csv("phenotype",index=False,sep=" ")
            
            call(["plink", "--bfile", gwas_file_name, "--assoc", "--pheno", "phenotype", allow-no-sex, "--geno", "0.00001", "--out", output_name])
        else:
            new_pheno.columns = ['FID','IID','Y']
            
            new_pheno.loc[(new_pheno['Y']==1),"Y"]=2
            new_pheno.loc[(new_pheno['Y']==0),"Y"]=1
            
            new_pheno.to_csv("phenotype",index=False,sep=" ")
            
            call(["plink", "--bfile", gwas_file_name, "--assoc", "--pheno", "phenotype", allow-no-sex, "--geno", "0.00001", "--out", output_name])
        
        result = pd.read_table(f"{output_name}.assoc",delimiter=r"\s+")
        
        return result   
    
    
    
    
    def qassoc(self, gwas_file_name, pheno , allow_no_sex = True,output_name="result"):
        fid = pd.DataFrame(pheno.iloc[:,0])
        new_pheno = pd.concat([fid,pheno],1)
        
        if sum(new_pheno.columns == ['FID','IID','Y']) == 3:
        
                
            
            new_pheno.to_csv("phenotype",index=False,sep=" ")
            
            call(["plink", "--bfile", gwas_file_name, "--linear", "--pheno", "--all-pheno", "phenotype", allow-no-sex, "--geno", "0.00001", "--out", output_name])
        else:
            new_pheno.columns = ['FID','IID','Y']
            
            
            new_pheno.to_csv("phenotype",index=False,sep=" ")
            
            call(["plink", "--bfile", gwas_file_name, "--linear", "--pheno", "--all-pheno", "phenotype", allow-no-sex, "--geno", "0.00001", "--out", output_name])
        
        result = pd.read_table(f"{output_name}.qassoc",delimiter=r"\s+")
        
        return result  
    
    
    
    
    def model(self, gwas_file_name, pheno, allow_no_sex = True, output_name="result"):
        
        fid = pd.DataFrame(pheno.iloc[:,0])
        new_pheno = pd.concat([fid,pheno],1)
        
        if sum(new_pheno.columns == ['FID','IID','Y']) == 3:
            
                
            new_pheno.loc[(new_pheno['Y']==1),"Y"]=2
            new_pheno.loc[(new_pheno['Y']==0),"Y"]=1
                
            
            new_pheno.to_csv("phenotype",index=False,sep=" ")
            
            call(["plink", "--bfile", gwas_file_name, "--model", "--pheno", "phenotype", allow-no-sex, "--geno", "0.00001", "--out", output_name])
        else:
            new_pheno.columns = ['FID','IID','Y']
            
            new_pheno.loc[(new_pheno['Y']==1),"Y"]=2
            new_pheno.loc[(new_pheno['Y']==0),"Y"]=1
            
            new_pheno.to_csv("phenotype",index=False,sep=" ")
            
            call(["plink", "--bfile", gwas_file_name, "--model", "--pheno", "phenotype", allow-no-sex, "--geno", "0.00001", "--out", output_name])
        
        result = pd.read_table(f"{output_name}.model",delimiter=r"\s+")
        
        return result
    
    
    
    def clump(self, gwas_file_name, clump_file_name, output_name = "result"):
        
        ## ex) clump_file_name format is "clump_file_name.assoc"

        call(["plink", "--bfile", gwas_file_name, "--clump", clump_file_name, allow-no-sex, "--out", output_name])
        
        result = pd.read_table(f"{output_name}.clumped",delimiter=r"\s+")
        return result
    
    #def to_matrix()
```
